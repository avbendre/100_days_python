## Star Patterns in Python

Sure, here are the Python codes for the basic star patterns you mentioned:

**1. Single Row of Stars:**

```
for i in range(5):
    print("*", end=" ")
print()
```

**2. Triangle of Stars:**

```
for i in range(1, 6):
    print("*" * i)
print()
```

**3. Inverted Triangle of Stars:**

```
for i in range(5, 0, -1):
    print("*" * i)
print()
```

**4. Diamond of Stars:**

```
for i in range(1, 6):
    print(" " * (5 - i), end="")
    print("*" * (2 * i - 1))
for i in range(4, 0, -1):
    print(" " * (5 - i), end="")
    print("*" * (2 * i - 1))
print()
```

**5. Hollow Square of Stars:**

```
for i in range(4):
    if i == 0 or i == 3:
        print("*" * 4)
    else:
        print("*" + " " * 2 + "*")
print()
```

These codes use nested loops to print stars in the desired patterns. The outer loop controls the number of rows, and the inner loop controls the number of stars in each row.

Feel free to experiment with these examples by changing the number of stars or the spacing between them!
