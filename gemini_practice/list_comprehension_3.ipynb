**11. Flatten a nested list:**

```
nested_list = [[1, 2, 3], [4, 5], [6, 7, 8, 9]]
flat_list = [item for sublist in nested_list for item in sublist]
# Output: [1, 2, 3, 4, 5, 6, 7, 8, 9]
```

**Explanation:** The inner loop iterates through each sublist, and the outer loop iterates through elements in each sublist, adding them to the `flat_list`.

**12. Find the element with the highest score in a list of objects:**

```
data = [
    {"name": "Alice", "score": 85},
    {"name": "Bob", "score": 92},
    {"name": "Charlie", "score": 78}
]
highest_scorer = max(data, key=lambda x: x["score"])
# Output: {'name': 'Bob', 'score': 92}
```

**Explanation:** This uses `max` with a custom key function that extracts the "score" attribute from each object in the `data` list.

**13. Generate all possible combinations of two lists:**

```
list1 = [1, 2, 3]
list2 = ["a", "b", "c"]
combinations = [(a, b) for a in list1 for b in list2]
# Output: [(1, 'a'), (1, 'b'), (1, 'c'), (2, 'a'), (2, 'b'), (2, 'c'), (3, 'a'), (3, 'b'), (3, 'c')]
```

**Explanation:** This leverages nested loops to create all possible pairs by iterating through each element in `list1` and `list2`.

**14. Remove duplicate elements from a list while preserving order:**

```
my_list = [1, 2, 2, 3, 4, 3, 5, 6, 6]
unique_list = [my_list[i] for i in range(len(my_list)) if my_list[i] not in my_list[:i]]
# Output: [1, 2, 3, 4, 5, 6]
```

**Explanation:** This checks if the current element is already present in the list before the current index. If not, it's added to the `unique_list`.

**15. Create a list of prime numbers up to a certain limit:**

```
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

limit = 20
prime_numbers = [num for num in range(2, limit + 1) if is_prime(num)]
# Output: [2, 3, 5, 7, 11, 13, 17, 19]
```

**Explanation:** This defines a `is_prime` function and uses it inside the list comprehension to filter only prime numbers within the specified `limit`.
